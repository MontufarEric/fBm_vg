```python
df_fbm = pd.read_csv('df_fbm.csv')
```

```python
df5000 = df_fbm[df_fbm['size']==5000]
df3000 = df_fbm[df_fbm['size']==3000]
df10000 = df_fbm[df_fbm['size']==10000]
df20000 = df_fbm[df_fbm['size']==20000]
```

```python
df5000G = df5000.groupby('H').mean().reset_index()
df3000G = df3000.groupby('H').mean().reset_index()
df10000G = df10000.groupby('H').mean().reset_index()
df20000G = df20000.groupby('H').mean().reset_index()
```

```python
fig=plt.figure()
plt.style.use('seaborn-darkgrid')
axes=fig.add_axes([0.1,0.1,2,2])

plt.plot(df5000G['H'],df5000G['slope'])
plt.plot(df3000G['H'],df3000G['slope'])
plt.plot(df10000G['H'],df10000G['slope'])
plt.plot(df20000G['H'],df20000G['slope'])
```

```python
fig=plt.figure()
plt.style.use('seaborn-darkgrid')
axes=fig.add_axes([0.1,0.1,2,2])

plt.plot(df5000G['H'],df5000G['intercept'])
plt.plot(df3000G['H'],df3000G['intercept'])
plt.plot(df10000G['H'],df10000G['intercept'])
plt.plot(df20000G['H'],df20000G['intercept'])
```

```python
fig=plt.figure()
plt.style.use('seaborn-darkgrid')
axes=fig.add_axes([0.1,0.1,2,2])
leg=['N = 3000', 'N = 5000', 'N = 10000','N = 20000',]

plt.plot(df3000G['H'],np.log(df3000G['intercept']),'-*')
plt.plot(df5000G['H'],np.log(df5000G['intercept']),'-*')
plt.plot(df10000G['H'],np.log(df10000G['intercept']),'-*')
plt.plot(df20000G['H'],np.log(df20000G['intercept']),'-*')

plt.legend(leg,fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('H',fontsize = 20)
plt.ylabel('Intercept (Log)',fontsize = 20)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df3000G['H'],np.log(df3000G['intercept'])
```

```python
df_3k_linear = df3000G[df3000G['H']>.35]
df_5k_linear = df5000G[df5000G['H']>.35]
df_10k_linear = df10000G[df10000G['H']>.35]
df_20k_linear = df20000G[df20000G['H']>.35]
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df_3k_linear['H'],np.log(df_3k_linear['intercept']))

x = np.linspace(0.37,1.03,1000)
y = x*slope + intercept
```

```python
fig=plt.figure()
plt.style.use('seaborn-darkgrid')
axes=fig.add_axes([0.1,0.1,2,2])

plt.plot(df_3k_linear['H'],np.log(df_3k_linear['intercept']),'-*')
plt.plot(x,y)

leg = ['Data','Adjusted Curve']

plt.legend(leg,fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('H',fontsize = 20)
plt.ylabel('Intercept (Log)',fontsize = 20)

print('r^2 = ', r_value)
```
